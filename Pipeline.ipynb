BASIC PIPELINE 

```python
#Pipeline In ML 
import pandas as pd 
import numpy as np
```

```python
d1 = {'Social_media_followers':[1000000, np.nan, 2000000, 1310000, 1700000, np.nan, 4100000, 1600000, 2200000, 1000000],
'Sold_out':[1,0,0,1,0,0,0,1,0,1]}
```

```python
df1 = pd.DataFrame(data = d1)
```

```python
df1
```

```python
#checking fornull
df1.isnull().sum()
```

```python
#Splitting data into X and y
X1 = df1[['Social_media_followers']]
```

```python
y1 = df1[['Sold_out']]
```

```python
#Split using train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Splitting into train and test
X1_train , X1_test , y1_train, y1_test = train_test_split(X1 , y1 , test_size= 0.3 , random_state=19 )
```

```python
X1_train.shape
```

```python
#adding an imputer 
#to fill in missing values
from sklearn.impute import SimpleImputer
```

```python
#fill in with mean
imputer = SimpleImputer(strategy='mean')
```

```python
#imprting model
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
#importing pipeline
from sklearn.pipeline import make_pipeline
```

```python
#fixing our imputer and LR_model in pipeline
pipe1 = make_pipeline(imputer , lr)
```

```python
pipe1
```

```python
pipe1.fit(X1_train, y1_train)
```

```python
pipe1.score(X1_train, y1_train)
```

```python
pipe1.named_steps.simpleimputer.statistics_
```

```python
pipe1.named_steps.logisticregression.coef_
```

ADVANCE PIPELINE 
With categorical and numerical pipeline

```python
d2 = {'Genre':['Rock', 'Metal', 'Bluegrass', 'Rock', np.nan, 'Rock', 'Rock', np.nan, 'Bluegrass', 'Rock'],
'Social_media_followers':[1000000, np.nan, 2000000, 1310000, 1700000, np.nan, 4100000, 1600000, 2200000, 1000000],
'Sold_out':[1,0,0,1,0,0,0,1,0,1]}
```

```python
df = pd.DataFrame(data= d2)
```

```python
d2
```

```python
df
```

```python
#split between X and Y  using the iloc appraoch
X = df.iloc[:, 0:2]
```

```python
y = df.iloc[:, 2]
```

```python
y
```

```python
#spitting using the train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= 0.3 , random_state= 70)
```

```python
#categorising columns into numerical and categorical data
num_cols = df['Social_media_followers']
cat_cols = df['Genre']
```

```python
#importing pipeline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
#creating numerical pipeline
num_pipeline = Pipeline(
    steps=[
        ('imputer' , SimpleImputer(strategy='mean')),
        ('scaler' , StandardScaler())
    ]
)
```

```python
#import one_hot encoder
from sklearn.preprocessing import OneHotEncoder
```

```python
#creating cat pipeline
cat_pipeline = Pipeline(
    steps= [
        ('impute' , SimpleImputer(strategy='most_frequent')),
        ('onehotencoder' , OneHotEncoder(handle_unknown='ignore' , sparse_output=False))
    ] 
)
```

```python
#using the column transformer 
from sklearn.compose import ColumnTransformer
```

```python
col_transformer = ColumnTransformer(
    transformers=[
        ('num_pipeline' , num_pipeline , num_cols),
        ('cat_pipeline' , cat_pipeline , cat_cols)
    ],
    remainder= 'drop',
    n_jobs= -1
)
```

```python
#importing decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
pipefinal = make_pipeline(col_transformer , dtc)
```

```python
pipefinal.fit(X_train, y_train)
```
